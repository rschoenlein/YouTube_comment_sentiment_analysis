```python
polarity = []

for comment in comments['comment_text']:
    try:
        polarity.append(TextBlob(comment).sentiment.polarity)
    except:
        polarity.append(0)
    
```

```python
len(polarity)
```

```python
comments['polarity'] = polarity
```

```python
comments.head(10)
```

```python
filter1 = comments['polarity']==1
```

```python
positive_comments = comments[filter1]
```

```python
filter2 = comments['polarity']==-1
```

```python
negative_comments = comments[filter2]
```

```python
!pip install wordcloud
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
total_positive_comments = ' '.join(positive_comments['comment_text'])
```

```python
positive_comments_cloud = WordCloud(stopwords=set(STOPWORDS)).generate(total_positive_comments)
```

```python
plt.imshow(positive_comments_cloud)
plt.axis('off')
```

```python
total_negative_comments = ' '.join(negative_comments['comment_text'])
```

```python
negative_comments_cloud = WordCloud(stopwords=set(STOPWORDS)).generate(total_negative_comments)
```

```python
plt.imshow(negative_comments_cloud)
plt.axis('off')
```
